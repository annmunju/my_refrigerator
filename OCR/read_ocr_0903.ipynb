```python
# pip install git+https://github.com/JaidedAI/EasyOCR.git
# pip install easyocr

import easyocr
import os

# use CPU
os.environ["CUDA_VISIBLE_DEVICES"]="-1"

# import sample
import glob

dir = './sample/'
files = glob.glob(dir+'*.jpeg')

# open image
from PIL import Image
```

```python
reader = easyocr.Reader(['ko','en']) # this needs to run only once to load the model into memory
```

```python
file = files[0]
texts = []
for loc, text, perc in reader.readtext(file):
    texts.append(text)
```
